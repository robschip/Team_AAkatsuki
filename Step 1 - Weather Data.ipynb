```python
#Average temperatur zeile dazupacken!
df_clean.insert(3, "avg_temp", 0)
df_clean["avg_temp"] = df_clean["avg_temp"].astype(float)

#den index reseten auf 1,2...,n
df_clean.reset_index(drop=True, inplace=True)
```

```python
#average temperatur berechnen und einfuegen
for x in df_clean.index:
    if df_clean.at[x, "max_temp"] == (df_clean.at[x, "min_temp"]):
        df_clean.at[x, "avg_temp"] = df_clean.at[x, "max_temp"]
    else:
        df_clean.at[x, "avg_temp"] = 1/2 * (df_clean.at[x, "max_temp"] + df_clean.at[x, "min_temp"])
```

```python
#gibt dir anzahl der tage des monats
def tagepromonth(month):
    if month==1 or month==3 or month==5 or month==7 or month==8 or month==10 or month==12:
        return 31
    if month==4 or month==6 or month==9 or month==11:
        return 30
    if month==2:
        return 28
    
dates=[]

for monat in range(12):
    for tag in range(tagepromonth(monat+1)):
        for stunde in range(24):
            if monat<9:
                mon='0'+str(monat+1)
            else:
                mon=str(monat+1)
            if tag<9:
                day='0'+str(tag+1)
            else:
                day=str(tag+1)
            if stunde<=9:
                hour='0'+str(stunde)
            else:
                hour=str(stunde)
                
            date="2019-"+mon+"-"+day+" "+hour+':00:00'
            dates.append(date)
```

```python
#achtung ausfuehrung dauert ewig, aber ist auch nur einmal auszufuehrender code!
#es fehlen 952 messungen aufgrund von duplikaten der timestamps
fehlende=dates[:]
k=0
for i in range(len(df_clean)):
    for j in range(len(fehlende)):
        if df_clean.iloc[i,0]==fehlende[j]:
            fehlende.remove(fehlende[j])
            break
            
print(len(fehlende))
```

```python
ids_todrop = []
# for-schleife vergleicht "nach links *und* rechts" --> Erstes "date-duplikat"-Paar manuell aussortieren (drop id=0, weil "falsy case: max = min" eintritt)
ids_todrop.append(0)
for i in range(1,len(dft)-1):
    if dft["date_time"].iloc[i] == dft["date_time"].iloc[i+1]:
    #bei duplikaten: *mindestens* ein record mit maxtemp = mintemp
        temp_eq_1 = dft["max_temp"].iloc[i] == dft["min_temp"].iloc[i]
        temp_eq_2 = dft["max_temp"].iloc[i+1] == dft["min_temp"].iloc[i+1]
        if temp_eq_1 == temp_eq_2 and i > 0:
            # 1st record: wetter (^= precip) gleich wie vorher *oder* nachher ("plausibel")
            if dft["precip"].iloc[i] == dft["precip"].iloc[i-1] or dft["precip"].iloc[i] == dft["precip"].iloc[i+2]:
                # 2nd -> raus
                ids_todrop.append(i+1)
            # 2nd record: wetter (^= precip) gleich wie vorher *oder* nachher ("plausibel")
            elif (dft["precip"].iloc[i+1] == dft["precip"].iloc[i-1]) or (dft["precip"].iloc[i+1] == dft["precip"].iloc[i+2]):
                # 1st -> raus
                ids_todrop.append(i)
            else:
                ids_todrop.append(i+1)
        elif temp_eq_1 == True:
            ids_todrop.append(i)
        elif temp_eq_2 == True:
            ids_todrop.append(i)
```

```python
dft_clean = dft.copy()
dft_clean = dft_clean.drop(ids_todrop)
```
